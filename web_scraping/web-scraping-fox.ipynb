```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from bs4 import BeautifulSoup
import time
from datetime import datetime, timedelta
```

```python

def scrape_page(url):
    driver = webdriver.Chrome()  # Initialize your webdriver, use appropriate driver for your browser
    driver.get(url)

    try:
        # Scroll to the bottom of the page to load all initial content
        driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")
        time.sleep(3)  # Add a short delay to ensure all content is loaded
        
        # Click the "Show More" button twice
        for _ in range(9):
            show_more_button = WebDriverWait(driver, 10).until(
                EC.element_to_be_clickable((By.CSS_SELECTOR, 'footer .button.load-more.js-load-more'))
            )
            show_more_button.click()
            # Wait for new content to load
            time.sleep(3)  # Add a short delay to ensure all content is loaded
        
        # Get the updated HTML after clicking "Show More" twice
        updated_html = driver.page_source
        
        # Use BeautifulSoup to parse the updated HTML
        soup = BeautifulSoup(updated_html, 'html.parser')
        
        # Extract all the data you need from the updated soup object
        # Example:
        articles = soup.find_all('article', class_='article')
        article_details = []
        for article in articles:
            date = article.find('span', class_='time').text.strip()
            title = article.find('h4', class_='title').text.strip()
            description = article.find('p', class_='dek').text.strip()
            article_url = article.find('a')['href']
            image_link = article.find('img')['src']
            
            article_details.append({
                'date' : date,
                'title': title,
                'description': description,
                'article_url': article_url,
                'image_link': image_link,
            })
        
        return article_details
    
    finally:
        driver.quit()  # Make sure to close the browser after scraping
```

```python
# URL of the webpage
url1 = 'https://www.foxnews.com/category/science/planet-earth/oceans'

# Call the function to start scraping
scraped_data1 = scrape_page(url1)
```

```python
# Add 'category' column and concatenate the dataframes
df_oceans_cat['category'] = 'oceans'
df_climatechange_cat['category'] = 'climate change'

# Concatenate the dataframes
articles_library = pd.concat([df_oceans_cat, df_climatechange_cat], ignore_index=True)

# Display the combined dataframe
articles_library
```

```python
from datetime import datetime, timedelta

def convert_to_date(date_str):
    now = datetime.now()
    if 'ago' in date_str:
        days_ago = int(date_str.split()[0])
        date = now - timedelta(days=days_ago)
    else:
        date_str += f", {now.year}"  # Append the current year
        date = datetime.strptime(date_str, "%B %d, %Y")
        # Check if the date is in the future compared to the current date
        if date > now:
            date = date.replace(year=date.year - 1)  # Subtract one year
    return date.strftime("%Y-%m-%d")

# Assuming articles_library is your dataframe and 'date' is the column containing date strings
articles_library['new_date'] = articles_library['date'].apply(convert_to_date)
```

```python
articles_library = articles_library[['category','new_date', 'title', 'description', 'article_url', 'image_link']]
```

```python
articles_library = articles_library.rename(columns={'new_date':'date'})
articles_library
```
