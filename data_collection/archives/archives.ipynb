```python
# Importing Libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df2 = pd.read_excel('CoralBleaching.xlsx')
```

```python
# Writing a function to explore each dataset

def explore_dataset(df):
    # Checking the shape of the dataset
    print("Shape of the dataset:", df.shape)
    print("\n")
    
    # Displaying the first few rows of the dataset
    print("First few rows of the dataset:")
    display(df.head())
    print("\n")
    
    # Getting information about the dataset
    print("Information about the dataset:")
    display(df.info())
    print("\n")
    
    # Checking for duplicated values per column
    duplicated_values_per_column = df.duplicated().any()
    print("Duplicated values per column:")
    display(duplicated_values_per_column)
    print("\n")
    
    # Counting duplicated values per column
    count_duplicated_values_per_column = duplicated_values_per_column.sum()
    print("Number of duplicated values per column:", count_duplicated_values_per_column)
    print("\n")
    
    # Checking for missing values
    print("Missing values per column:")
    display(df.isnull().sum())
    print("\n")
    
    # Renaming columns to lowercase
    df.rename(columns=lambda x: x.lower(), inplace=True)
    
    # Displaying number of unique values per column
    print("Number of unique values per column: ")
    display(df.nunique())
```

```python
# Writing a function to handle missing values
def drop_columns_with_missing_values(df):
    """
    Drop columns with missing values from the DataFrame.

    Parameters:
    df (DataFrame): The DataFrame containing the data.

    Returns:
    None
    """
    # Drop columns with missing values in place
    df.dropna(axis=1, inplace=True)
```

```python
# Writing a function to replace column name
def replace_column_name(df, old_column_name, new_column_name):
    """
    Replace a column name in the DataFrame.

    Parameters:
    df (DataFrame): The DataFrame containing the data.
    old_column_name (str): The current name of the column to be replaced.
    new_column_name (str): The new name to assign to the column.

    Returns:
    None
    """
    # Replace the column name
    df.rename(columns={old_column_name: new_column_name}, inplace=True)
```

```python
# Writing a function to convert a column to datetime object :
def convert_to_datetime(df, columns):
    """
    Convert specified columns in the DataFrame to datetime objects.

    Parameters:
    df (DataFrame): The DataFrame containing the data.
    columns (list): A list of column names to convert to datetime.

    Returns:
    None
    """
    # Convert columns to datetime objects
    for col in columns:
        df[col] = pd.to_datetime(df[col])
```

```python
# Writing a function to check the percentage of missing values of each column :
def missing_percentage(df):
    """
    Calculate the percentage of missing values in each column of a DataFrame.
    
    Parameters:
    df (DataFrame): Input DataFrame.
    
    Returns:
    missing_percentages (Series): Series containing the percentage of missing values in each column.
    """
    # Calculate the total number of missing values in each column
    missing_values = df.isnull().sum()
    
    # Calculate the total number of values in each column
    total_values = df.shape[0]
    
    # Calculate the percentage of missing values in each column
    missing_percentages = (missing_values / total_values) * 100
    
    return missing_percentages
```

```python
# Let's chech the missing percentage for our second dataset : 
missing_percentages = missing_percentage(df2)
missing_percentages
```

```python
# Writing a function to fill in missing values with "na":
def replace_missing_with_na(df, columns):
    """
    Replace missing values in specified columns with "na".

    Parameters:
    - df (DataFrame): The DataFrame containing the data.
    - columns (list): List of column names to replace missing values with "na".

    Returns:
    - DataFrame: The DataFrame with missing values replaced by "na".
    """
    df[columns] = df[columns].fillna("na")
    return df

# Usage example
# Assuming df is your DataFrame and columns_with_missing is the list of columns with missing values
columns_with_missing = ['region', 'subregion', 'location']
df_with_na = replace_missing_with_na(df2, columns_with_missing)

# Check if missing values have been replaced
print(df_with_na.isnull().sum())
```

```python
# Writing a function to rename values to lowercase:
def rename_values_to_lower(df):
    """
    Rename all string/object values in the DataFrame to lowercase.

    Parameters:
    - df (DataFrame): The DataFrame containing the data.

    Returns:
    - DataFrame: The DataFrame with string/object values renamed to lowercase.
    """
    string_columns = df.select_dtypes(include=['object']).columns
    for column in string_columns:
        df[column] = df[column].str.lower()
    return df

# Usage example
# Assuming df is your DataFrame
df_lower = rename_values_to_lower(df2)

# Display the first few rows of the DataFrame to verify the changes
display(df_lower.head())
```

```python
# Handle missing values in the 'month' column (replace with -1)
df2['month'] = df2['month'].fillna(-1)

# Convert the 'month' column to integer type
df2['month'] = df2['month'].astype(int)

# Display the data types of the DataFrame to verify the changes
print(df2.dtypes)
```

```python
# Filter out rows with 'na' and 'severity unknown' in the 'bleaching_severity' column
filtered_df = df2[~df2['bleaching_severity'].isin(['severity unknown']) & ~df2['region'].isin(['na'])]

# Define the order of the categories
severity_order = ['no bleaching', 'low', 'medium', 'high']

# Convert the 'bleaching_severity' column to categorical with the specified order
filtered_df['bleaching_severity'] = pd.Categorical(filtered_df['bleaching_severity'], categories=severity_order, ordered=True)

# Group the filtered data by severity code and categorical variable (e.g., region) and normalize
severity_distribution = filtered_df.groupby(['region', 'bleaching_severity']).size().unstack(fill_value=0)
severity_distribution = severity_distribution.div(severity_distribution.sum(axis=1), axis=0) * 100

# Define custom color palette
pal = ["#9b59b6", "#e74c3c", "#34495e", "#2ecc71"]

# Plot a stacked bar chart with custom colors
severity_distribution.plot(kind='bar', stacked=True, figsize=(8, 5), color=pal, alpha=0.4)
plt.title('Distribution of Bleaching Severity Across Regions (Stacked, 100%)')
plt.xlabel('Region')
plt.ylabel('Percentage')
plt.xticks(rotation=45)  # Rotate x-axis labels for better readability
plt.legend(title='Bleaching Severity', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()  # Adjust layout to prevent clipping of labels
plt.show()
```
